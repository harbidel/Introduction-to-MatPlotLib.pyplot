# Introduction to MatPlotLib

```python
#importing the matplotlib library
import matplotlib.pyplot as plt
```

```python
#this allows us to be able to visualize our graphs
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# ploting graph of x and y
plt.plot(x, y,'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
# printing multiple plots using plt.subplot
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,4,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,4,2)
plt.plot(y, x, 'g*-');
plt.subplot(1,4,4)
plt.plot(x, y, 'y--') 
plt.subplot(1,4,3)
plt.plot(y, x, 'b*-');
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# printing many graphs on a single figure

# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes,[left,bottom, height width]
axes2 = fig.add_axes([0.6, 0.2, 0.25, 0.15])

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Axes is an array of axes to plot on
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_xlabel('y')
    ax.set_title('title')

# Display the figure object    
fig
```
